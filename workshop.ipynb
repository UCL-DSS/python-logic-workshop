### Conditional Statements
Sometimes you will only want code to run if a certain condition is met. To do this we use ````if````, ````else```` and ````elif````. These will require comparision operators examples of which are given below. At the end of any conditional statement we need a ````:````. Then code that runs in the event the condition is met is placed below but needs to start with four spaces. You can just press Tab to create four spaces although a lot of Python editors automatically create spaces for you. The four spaces are there to indicate the fact that this code is to run if the condition is met. 

`````python
# 1) If 1 is greater than 0 then 1 will be printed.
if 1 > 0: 
    print("1) True")

# 2) If 1 is greater than or equal to 0 print 0
if 1 >= 0: 
    print("2) True")
    
# 3) If 0 is less than 1 print 1.
if 0 < 1:
    print("3) True")
    
# 4) If 1 is less than or equal to 2 print 2
if 1 <= 2:
    print("4) True")
    
# 5) If 2 equals 2 print 2
if 3 == 3:
    print("5) True")
    
# 6) If 4 is not equal to 5 print 4
if 4 != 5:
    print("6) True")
    
# 7) If the string "L" can be found in the string "UCL"
if "L" in "UCL":
    print("7) True")
    
# 8) If the string "Q" can not be found in the string "UCL"
if "Q" not in "UCL":
    print("8) True")
    
# 9) If 8 is less than 9 but greater than 7
if 9 > 8 > 7:
    print("9) True")

# 10) Alternative way of writing 9)
if 9 > 8 and 8 > 7:
    print("10) True")
    
# 11) If 4 equals 5 or if 4 greater than 3
if 4 == 5 or 4 > 3:
    print("11) True")
`````

### Loops
Often you'll want to perform certain tasks multiple times. This is where loops come in! There are two types of loops that are commonly used. 

#### for Loops
When you have a list or a range of values you can loop over those using a ````for```` loop. Below is an example of such a loop. The first line ````for i in range(101):```` means that you create a variable ````i```` that runs over a range of values. Since we have ````range(101)```` the values of ````i```` will run from $0$ to $100$. This loop prints each ````i```` value to demonstrate how this variable changes over the loop.

`````python
# Loops over all integers from 0 to 100
for i in range(101):
    print(i)
`````

Note the colon at the end of the first line and the space left before the ````print```` statement. This is known as indentation and it simply tells to processer that the ````print```` statement lies within the loop. If we had no spaces before the ````print```` statement then the print statement would not run as part of the loop. Typically this requires four spaces which you'll find the Tab key useful for as this moves a line four spaces along.

`````python
# Loops over all integers from 0 to 100
for i in range(101):
print(i)
`````

Below we use a ````for```` loop to calculate the sum of all integers from $1$ to $100$ inclusive.

`````python
# Variable that will contain the sum 
total = 0

# Loop runs from 1 to 100
for x in range(1,101):
    # Adds current x value to the sum
    total = total + x

# Output result
print(total)   
`````

#### while Loops
These are loops that run while a certain condition is true. Below is the ````while```` loop equivalent of the ````for```` loop that calculates the sum of integers from $1$ to $100$.

`````python
# This variable acts as a counter for the loop
n = 1

# Variable that will contain the sum 
total = 0

# While n is less than or equal to 100 the loop runs
while n <= 100:
    # Adds current value of n to the sum
    total += n
    
    # Moves n to the next value (IF THIS LINE WASN'T HERE THE LOOP WOULD RUN INFINITELY!)
    n += 1
    
    # For code to be part of the loop it has to be correctly indented like this comment
# Whereas this comment is outside the loop! 
    
# Outputting the result 
print(total)
`````

Be aware that your loops should contain exit conditions. For the ````while```` loop above it is very easy to forget the line ````n += 1````. This line increases the value of ````n```` by $1$ each cycle so that after $100$ cycles ````n```` is greater than ````100```` so the loop condition for ````n```` being less than or equal to $100$ is no longer true. If you omitted the line ````n += 1```` then the loop condition would always be true so the loop would keep running indefinetely!

#### Nesting loops 
One can create a loop within another as can be seen below. The example loops over all the terms in one of strings that was split into a list above and loop over the characters in each term printing them out individually. Also note the individual loops are slightly different. Rather than running over a range the loops below run over the split string or string. The line
````for i in x:```` runs over each term in ````x````. Here ````i```` varies by starting of being defined as the first term in ````x```` and then moves through each term. The second loop runs each over each term in the current ````i````. As you can imagine if you had lots of nested loops that may start to prove computationally taxing. With this in mind it is often preferable to avoid too many complicated loop structures to try an improve the efficiency of your code.

`````python
# Defining x as a specific string
x = "One does not simply walk into ExCel."

# Redifines x as a list of the words in the string x was originally defined as above
x = x.split()

# Loops over each item in the list where i is the current term
for i in x:
    # Loops over each character in i
    for j in i:
        print(j)
`````

`````python
# Defines a function sum_n which passes an argument nmax
def sum_n(nmax):
    '''This is the doctring of the function. Here you write what the function
    is supposed to do so that users know what it does and what arguments need
    to be passed. You can check the doctring of a predefined function by 
    pressing shift+tab while inside the function'''
    
    # This variable acts as a counter for the loop
    n = 1

    # Variable that will contain the sum 
    value = 0

    # While n is less than or equal to 100 the loop runs
    while n <= nmax:
        # Adds current value of n to the sum
        value += n
    
        # Moves n to the next value 
        n += 1
    
    # Returns the total
    return value
`````

#### Global and Local Variables
When defining a function it is important to be aware of what is a local variable and what is global. Any variable defined inside the function will only exist while the function runs. This is a local variable as it is "local" to the function. To change a function to a global variable we use a ````global```` line where we write all the variables we want to be global variable. This means that the function will change the existing variables or create new variables in the ````global```` line that will exist when the function is finished running. ## Optional (Object Oriented)
If you want to create a new type of object that has certain attributes or functions that apply specifically to that type of object then you can do this by creating a new ````class````. A type of object is known as a class while a single instnce of that class is considered to be an object. A simple example of class already discussed could be a string. If we have a string object defined ````string```` then using ````string.split()```` applies a method to the string that as we saw splits the string into a list of terms wherever there is a space seperating characters in the string. Later in the workshops we will encounter algorithms from Sci-Kit Learn that are objects with a number of methods assoiciated with them in order to train, test and make predictions.

If we wish to start making our own objects then we begin by specifying ````class object_name:````. Then indented within this structure we can define various functions. What is important to bare in mind is that these functions are methods associated with your object rather than functions you call like the ones above. There are a few specific reserved methods which when defined have a rather specific behaviour. An example of this is ````__init__```` which is intended to control the setup of a new object. The line ````def __init__(self, x):```` defines a method that will run when we create a new object of this class without needing to call this method. Here the argument ````self```` refers to our object itelf and ````input_list```` refers to an additional argument we call when creating an object. Here we wish to take a list as our ````input_list```` so that we can apply various methods to it later. Within the ````__init__```` method we define various attributes. In this example we define the list itself as an attribute of the object with ````self.values = input_list````, where ````values```` is our attribute and ````input_list```` is the list we take as an argument when creating our object. The second attribute given is the length of the list through the line ````self.len = len(input_list)````, where the ````len```` function returns the length of the list.  It's important to note that the list is not the object in this example. Instead it is an attribute of our object.

`````python
# Creates a new type of object
class my_list:
    # Method/function that runs when you call your object
    def __init__(self, input_list):
        # Attribute for values in the list
        self.values = input_list
        
        # Attribute for length of the list
        self.len = len(input_list)
`````

Below we create an object of the same class we created in the code cell above. The ````__init__```` method runs automatically so we need to provide our ````input_list```` argument. Here we use ````[1,0,1]````. We define our new object to be called ````x````. We can then call attributes of ````x```` that we defined when making our class by calling ````x.values```` and ````x.len```` to return their respective attributes.

`````python
# Creates an object of the same class as my_list
x = my_list([1,0,1])

# Returns the list we defined in our new object
print(x.values)

# Returns the length of the list defined in our object
print(x.len)
`````

Below we recreate our class in the same way as before this time adding a new method. Our new method shall be defined with ````def summation(self):````. The method is intented to find the total sum of all values in the list attributed to our object. It works by using a for loop over the length of the list slicing for the $i^{th}$ value in the list for each loop cycle to add that value to the list total.

`````python
# Creates a new type of object
class my_list:
    # Method that runs when you call your object
    def __init__(self, input_list):
        # Attribute for values in the list
        self.values = input_list
        
        # Attribute for length of the list
        self.len = len(input_list)
    
    # Method that will return the sum of the list
    def summation(self):
        # Sum total
        total = 0
        
        # Loops over values of list in our object
        for i in range(self.len):
            # Adds current list value to sum total
            total = total + self.values[i]
        
        # Returns sum total
        return total           
`````

Defining ````x```` with the same list as we did previously then we apply our ````summation```` method to the object. We do the same for another object ````y````, which we will provide the list ````[9,3,4,7]```` as input.

`````python
# Creates an object of the same class as my_list
x = my_list([1,0,1])

# Uses the summation method to find total of the values in the list
print(x.summation())

# Creates an object of the same class as my_list
y = my_list([9,3,4,7])

# Uses the summation method to find total of the values in the list
print(y.summation())
`````
